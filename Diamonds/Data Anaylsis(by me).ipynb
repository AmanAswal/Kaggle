```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df= pd.read_csv('diamonds.csv')
print(df.shape)
df.head()
```

```python
df1 = df.copy()
```

```python
df1.drop(columns=['Unnamed: 0'], axis=1, inplace=True )
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.isnull().sum()
```

# Q1. Does color of diamond effects its price?

```python
plt.figure(figsize=(8,5))
sns.countplot('color', data=df1)
plt.title('No. of diamonds w.r.t Color')
```

#### color 'G' has max frequency

### Now, plot relationship between color and price of diamond

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
df1.groupby('color').price.mean().plot.bar()
plt.title('Price v/s color of diamond', fontsize=15)
plt.ylabel('Price', fontsize=15)
plt.xlabel('Color', fontsize=15)
plt.xticks(rotation=0, fontsize=12)

plt.subplot(1,2,2)
df1.groupby('color').price.mean().plot()
plt.title('Price v/s color of diamond', fontsize=15)
plt.ylabel('Price', fontsize=15)
plt.xlabel('Color', fontsize=15)
```

#### Yes, the price of diamond is not same for every color.
#### color 'J' is the most expensive diamond

## 1. Carat and color

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
sns.distplot(df1['carat'],kde=False,bins=25)
plt.title('No. of diamonds w.r.t Carat')

plt.subplot(1,2,2)
sns.countplot('color', data= df1)
plt.title('No. of Diamonds w.r.t Color')
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
df1.groupby('color').carat.mean().plot.bar()
plt.title('Carat v/s Color of diamond', fontsize=15)
plt.ylabel('Carat', fontsize=15)
plt.xlabel('Color', fontsize=15)
plt.xticks(rotation=80, fontsize=12)

plt.subplot(1,2,2)
df1.groupby('color').carat.mean().plot()
plt.title('Carat v/s Color of diamond', fontsize=15)
plt.ylabel('Carat', fontsize=15)
plt.xlabel('Color', fontsize=15)
```

#### observation: Slope is increasing. Color of diamond changes with different carat value.

## 3. Carat and Clarity
